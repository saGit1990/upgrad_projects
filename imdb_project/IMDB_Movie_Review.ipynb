-  ### Subtask 2.3: Drop unecessary rows using columns with high Null percentages

Now, on inspection we noticed that some columns have large percentage (greater than 5%) of Null values. Drop all the rows which have Null values for such columns.

```python
# dropping some rows
df = df[df.isnull().sum(axis = 1) <= 5]
100 * df.isnull().sum() / len(df.index)
```

-  ### Subtask 2.4: Drop unecessary rows

Some of the rows might have greater than five NaN values. Such rows aren't of much use for the analysis and hence, should be removed.

```python
# dropping the rows here
df = df[~np.isnan(df['gross'])]
df = df[~np.isnan(df['budget'])]
100 * df.isnull().sum() / len(df.index)
```

-  ### Subtask 2.5: Fill NaN values

we might notice that the `language` column has some NaN values. Here, on inspection, you will see that it is safe to replace all the missing values with `'English'`.

```python
# Write your code for filling the NaN values in the 'language' column here
df.language = df.language.fillna('English')
100 * df.isnull().sum() / len(df.index)
```

## Task 3: Data Analysis

-  ### Subtask 3.1: Change the unit of columns

Converting the unit of the `budget` and `gross` columns from `$` to `million $`.

```python
#  code for unit conversion here
df.budget = (0.000001 * df.budget)
df.gross = 0.000001 * df.gross  
df.head()
```

-  ### Subtask 3.2: Find the movies with highest profit

    1. Creating a new column called `profit` which contains the difference of the two columns: `gross` and `budget`.
    2. Sorting the dataframe using the `profit` column as reference.
    3. Extracting the top ten profiting movies in descending order and store them in a new dataframe - `top10`

```python
# code for creating the profit column here
df['profit'] = df.gross - df.budget
df.head()
```

```python
#  code for sorting the dataframe here
df = df.sort_values(by = ['profit'], ascending = False)
```

```python
top10 = df[0:10]

# Code to get the top 10 profiting movies here
top10
```

-  ### Subtask 3.3: Drop duplicate values

After we found out the top 10 profiting movies, we have notice a duplicate value. So, it seems like the dataframe has duplicate values as well. Dropping the duplicate values from the dataframe and repeat `Subtask 3.2`.

```python
# Code for dropping duplicate values here
df = df.drop_duplicates()
df.shape
```

```python
# Code for repeating subtask 2 here
df['profit'] = df.gross - df.budget
df = df.sort_values(by = ['profit'], ascending = False)
df.head()
```

-  ### Subtask 3.4: Find IMDb Top 250

    1. Creating a new dataframe `IMDb_Top_250` and store the top 250 movies with the highest IMDb Rating (corresponding to the column: `imdb_score`). Also making sure that for all of these movies, the `num_voted_users` is greater than 25,000.
Also add a `Rank` column containing the values 1 to 250 indicating the ranks of the corresponding films.
    2. Extracting all the movies in the `IMDb_Top_250` dataframe which are not in the English language and store them in a new dataframe named `Top_Foreign_Lang_Film`.

```python
# Code for extracting the top 250 movies as per the IMDb score here. Make sure that you store it in a new dataframe 
# and name that dataframe as 'IMDb_Top_250'
IMDb_Top_250 = df[df.num_voted_users>25000].sort_values(by = ['imdb_score'], ascending = False)[0:250]
IMDb_Top_250['Rank'] = df.imdb_score.rank(ascending = False).astype(int)
IMDb_Top_250.shape
```

```python
Top_Foreign_Lang_Film = IMDb_Top_250[IMDb_Top_250.language != 'English']
Top_Foreign_Lang_Film.loc[Top_Foreign_Lang_Film.movie_title.str.contains('Veer') ]
```

- ### Subtask 3.5: Find the best directors

    1. Groupping the dataframe using the `director_name` column.
    2. Sorting out the top 10 directors for whom the mean of `imdb_score` is the highest and store them in a new dataframe `top10director`. 

```python
#  code for extracting the top 10 directors here
top10director = df.groupby(by = ['director_name'])['imdb_score'].mean().sort_values(ascending = False)[0:10]
top10director
```

-  ### Subtask 3.6: Find popular genres

```python
# code for extracting the first two genres of each movie here
new = df.genres.str.split('|', expand = True)
df['genre_1'] = new[0]
df['genre_2'] = new[1]
df.genre_2 = df.genre_2.fillna(value = df.genre_1)
df.head()
```

-  ### Subtask 3.7: Find the critic-favorite and audience-favorite actors

    1. Create three new dataframes namely, `Meryl_Streep`, `Leo_Caprio`, and `Brad_Pitt` which contain the movies in which the actors: 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' are the lead actors. Use only the `actor_1_name` column for extraction. Also, make sure that you use the names 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' for the said extraction.
    2. Append the rows of all these dataframes and store them in a new dataframe named `Combined`.
    3. Group the combined dataframe using the `actor_1_name` column.
    4. Find the mean of the `num_critic_for_reviews` and `num_user_for_review` and identify the actors which have the highest mean.

```python
#  code for creating three new dataframes here
Meryl_Streep = df[df.actor_1_name == 'Meryl Streep']
```

```python
Leo_Caprio = df[df.actor_1_name == 'Leonardo DiCaprio']
```

```python
Brad_Pitt = df[df.actor_1_name == 'Brad Pitt']
```

```python
# code for combining the three dataframes here
Combined = pd.concat([pd.concat([Meryl_Streep,Leo_Caprio]) , Brad_Pitt])
```

```python
#  code for grouping the combined dataframe here
Combined_Actors = Combined.groupby(by = ['actor_1_name'])
```

```python
#  code for finding the mean of critic reviews and audience reviews here
Combined_Actors['num_critic_for_reviews' ,'num_user_for_reviews' ].mean().sort_values(by = ['num_critic_for_reviews','num_user_for_reviews'],ascending = False)
```
