```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()



q1 = lead_df['TotalVisits'].quantile(0.25)
q3 =  lead_df['TotalVisits'].quantile(0.75)
iqr1 = q3 - q1
lead_df = lead_df[(lead_df['TotalVisits']>=q1-1.5*iqr1) & (lead_df['TotalVisits']<=q3+1.5*iqr1)]
```

```python
x=lead_df.drop(['Prospect ID','Lead Number','Converted'],axis=1)
y=lead_df['Converted']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , train_size = 0.7 , test_size = 0.3 , random_state = 100)
```

```python
x_train[num_list] = scaler.fit_transform(x_train[num_list])
```

```python
#logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import statsmodels.api as sm
LR1 = sm.GLM(y_train , (sm.add_constant(x_train)), family = sm.families.Binomial())
LR1.fit().summary()
```

```python
# PCA adn Incremental PCA

from sklearn.decomposition import PCA
from sklearn.decomposition import IncrementalPCA

pca = PCA(svd_solver = 'randomized' , random_state = 50)
pca.fit(x_train)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
colnames = list(x_train.columns)
df_pcs =pd.DataFrame({'PC1': pca.components_[0] , 'PC2' : pca.components_[1] , 'Features' :colnames})
df_pcs.head()
```

```python
plt.scatter(df_pcs.PC1 , df_pcs.PC2)
plt.xlabel('PC1')
plt.ylabel('PC2')

for i , txt in enumerate(df_pcs.Features):
    plt.annotate(txt, (df_pcs.PC1[i],df_pcs.PC2[i]))
plt.show()
```

```python
# Screeplot

fig = plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('components')
plt.ylabel('variance')
plt.show()

# about 90% variables can be explained by 30 variables
```

```python
# Incremental PCA
pca_final = IncrementalPCA(n_components = 33)
```

```python
df_train_pca = pca_final.fit_transform(x_train)
df_train_pca = pd.DataFrame(df_train_pca)
df_train_pca.shape
```

```python
# correlation plot

corrmat = np.corrcoef(df_train_pca.transpose())
plt.figure(figsize=(23,20))
sns.heatmap(corrmat , annot = True)
plt.show()
```

```python
# Print minimum and maximum correlation values

corrmat_nodiag = corrmat - np.diagflat(corrmat.diagonal())
corrmat_nodiag.max() , corrmat_nodiag.min()
```

```python
# Test data

df_test_pca = pca_final.transform(x_test)
df_test_pca.shape  # array need to be converted to DF
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

logreg = LogisticRegression()
model = logreg.fit(df_train_pca , y_train)
```

```python
model_pred = model.predict_proba(df_test_pca)[:,1]
"{:2.2}".format(metrics.roc_auc_score(y_test, model_pred)) , model_pred
```
